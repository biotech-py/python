### Importing Pandas

```python
import pandas as pd
data = [ 100, 102, 104]
series = pd.Series(data)
print(series)
```

### Custom Indexing and Loc 

```python
import pandas as pd
data = [ 100, 102, 104]
series = pd.Series(data, index=["a","b","c"])
print(series)
print(series.loc["b"])
series.loc["c"]=200
print(series)
```

### Filter by value

```python
data=[100,102,104,200,202]
series=pd.Series(data,index=["a","b","c","d","e"])
print(series[series>=200])
```

```python
calories={"Day 1": 1750, "Day 2": 2100, "Day 3": 1700}
series=pd.Series(calories)
print(series)
```

```python
series=pd.Series(calories)
series.loc["Day 3"]+=500
print(series)
```

```python
series=pd.Series(calories)
print(series[series<2000])
```
