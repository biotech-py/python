### Importing Pandas

```python
import pandas as pd
data = [ 100, 102, 104]
series = pd.Series(data)
print(series)
```

### Custom Indexing and Loc 

```python
import pandas as pd
data = [ 100, 102, 104]
series = pd.Series(data, index=["a","b","c"])
print(series)
print(series.loc["b"])
series.loc["c"]=200
print(series)
```

### Filter by value

```python
data=[100,102,104,200,202]
series=pd.Series(data,index=["a","b","c","d","e"])
print(series[series>=200])
```

```python
calories={"Day 1": 1750, "Day 2": 2100, "Day 3": 1700}
series=pd.Series(calories)
print(series)
```

```python
series=pd.Series(calories)
series.loc["Day 3"]+=500
print(series)
```

```python
series=pd.Series(calories)
print(series[series<2000])
```

### Data Frame

```python
import pandas as pd
data = { "Name": ["Nirupam", "Nilanjan", "Nibir"],
        "Age": [22, 23, 50]
}
df=pd.DataFrame(data, index=["1","2","3"])
print(df)
```

```python
df=pd.DataFrame(data, index=["Employee 1", "Employee 2", "Employee 3"])
print(df.loc["Employee 2"])
```

```python
# Adding a new column


df["Job"] = ["N/A", "Product Designer", "Manager"]
print(df)
```

```python
# Adding a new row

new_row = pd.DataFrame([{"Name" : "Mrigalekha" , "Age" : 48 , "Job" : " Teacher" }], index = ["Employee 4"])
df = pd.concat([df, new_row])
print(df)
```

### Adding new rows

```python
new_rows = pd.DataFrame([{"Name" : "Mrigalekha" , "Age" : 48 , "Job" : " Teacher" },
                         { "Name" : "Gita" , "Age" : 100 , "Job" : " Bed Rest"}], index = ["Employee 4", "Employee 5"])
df = pd.concat([df, new_rows])
print(df)
```
