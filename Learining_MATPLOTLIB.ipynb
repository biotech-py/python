###                               MATPLOTLIB

```python
# Importing the matplotlib library and printing its version

import matplotlib
import matplotlib.pyplot as plt
print(matplotlib.__version__)
```

#### Line Plot

```python
x = [ 2023, 2024, 2025, 2026]
y = [15, 25, 30, 20]
plt.plot(x, y)

plt.show()
```

```python
# Using NumPy arrays to create a line plot

import numpy as np
x = np.array([2023, 2024, 2025, 2026])
y = np.array([15, 25, 30, 20])
plt.plot(x, y)
plt.show()  
```

#### Plot Customization

```python
# Single Line Plot with Custom Markers and Colors

plt.plot(x, y, marker='o', markersize=8, markerfacecolor='red', markeredgecolor='black', linestyle='-', linewidth=2, color='blue')
plt.show()
```

```python
# Multiple Line Plots with Custom Styles


x= [2023, 2024, 2025, 2026]
y1 = [15, 55, 30, 80]
y2 = [10, 33, 25, 50]
line_style = dict(marker='o', markersize=8, markerfacecolor='red', markeredgecolor='black', linestyle='-', linewidth=2)
plt.plot(x, y1, color='blue', **line_style)
plt.plot(x, y2, color='orange', **line_style)
plt.show()
```

```python
# Customizing Plot Appearance with Titles, Labels, and Tick Parameters

x= [2023, 2024, 2025, 2026]
y1 = [15, 55, 30, 80]
y2 = [10, 33, 25, 50]
y3 = [5, 20, 15, 40]
y4 = [8, 25, 18, 45]
plt.title("Class Size", fontsize=14, family='Arial', fontweight='bold', color="#ff0000")
plt.xlabel("Year", fontsize=12, family='serif', fontweight='bold', color='#2dbefc')
plt.ylabel("Number of Students", fontsize=12, family='serif', fontweight='bold', color='#2dbefc')
plt.tick_params(axis='both',color='brown')
plt.plot(x, y1, marker= '.')
plt.plot(x, y2, marker= 'o')
plt.plot(x, y3, marker= 's')
plt.plot(x, y4, marker= '^')
plt.xticks(x)
plt.show()
```

```python
# Customizing Grid Lines in a Plot

x=[1,2,3,4,5]
y=[5,10,15,20,25]
plt.grid()
plt.grid(axis='x', color='red', linestyle='--', linewidth=1)
plt.grid(axis='y', color='blue', linestyle='-.', linewidth=1)
plt.plot(x, y)
plt.show()
```

#### Bar Chart

```python
# Creating a Horizontal Bar Chart with Custom Colors and Labels

category = ["Grains", "Vegetables", "Fruits", "Dairy", "Meat", "Beverages", "Snacks", "Frozen Foods", "Household Items"]
values = [20, 15, 25, 10, 30, 18, 12, 8, 22]
plt.barh(category, values, color='red')
plt.title(" Daily Consumption of Food Categories", fontsize=14, family='Arial', fontweight='bold', color="#ff0000")
plt.xlabel(" Food", fontsize=12, family='serif', fontweight='bold', color='#2dbefc')
plt.ylabel(" Quantity", fontsize=12, family='serif', fontweight='bold', color='#2dbefc')
plt.tight_layout()
plt.show()
```

#### Pie Chart

```python
# Creating a Pie Chart with Custom Colors, Labels, and Exploded Slices

category = [ "Freshmen", "Sophomores", "Juniors", "Seniors"]
values = [100, 150, 120, 80]
color = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
plt.pie(values, labels=category, colors=color, autopct='%1.1f%%', startangle=90, explode=[0, 0, 0, 0.1], shadow=True)
plt.title("Bro Code College")
plt.show()
```

#### Scatter Plot

```python
x1 = [ 0,1,1,2,3,4,5,6,6,7,8]
y1 = [55,60,65,62,70,75,80,78,85,90,95]

x2 = [ 0,1,2,2,3,4,5,6,7,8,9]
y2 = [45, 63, 54, 68, 72, 78, 82, 80, 88, 92, 94]
plt.scatter(x1, y1, color='blue', label='Class A', alpha=1, s = 100)
plt.scatter(x2, y2, color='orange', label='Class B', alpha=1, s = 100)
plt.title("Test Scores ")
plt.xlabel("Hours Studied")
plt.ylabel("Grade")
plt.legend()
plt.show()
```

#### Histogram

```python
scores = np.random.normal(loc=80, scale=10, size=100)
scores = np.clip(scores, 0, 100)
plt.hist(scores, bins=10, color='purple', edgecolor='black')
plt.title("Exam Scores")
plt.xlabel("Score")
plt.ylabel("# of students")
plt.show()
```

#### Subplot

```python
x=np.array([1,2,3,4,5])
figure, axes = plt.subplots(2, 2)
axes[0, 0].plot(x, x*2, color='blue')
axes[0, 0].set_title(" x*2 ")
axes[0, 1].plot(x, x**2, color='orange')
axes[0, 1].set_title(" x**2 ")
axes[1, 0].plot(x, x**3, color='green')
axes[1, 0].set_title(" x**3 ")
axes[1, 1].plot(x, x**4, color='red')
axes[1, 1].set_title(" x**4 ")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("pokemon.csv")
type_counts = df['Type1'].value_counts(ascending=True)
plt.barh(type_counts.index, type_counts.values, color='firebrick', edgecolor='black')
plt.title("# of Pokemon by Primary Type")
plt.xlabel("Count")
plt.ylabel("Type")
plt.show()  
```
