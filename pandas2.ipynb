### Filtering using Python

```python
# Filtering Tall pokemons of Height > = 2

import pandas as pd
df = pd.read_csv("pokemon.csv")
tall_pokemon = df[df["Height"]>=2]
print(tall_pokemon)
```

```python
# Filtering Heavy pokemons of Weight > 100

heavy_pokemon = df[df["Weight"]>100]
print(heavy_pokemon)
```

```python
# Filtering water type pokemons from Type 1

water_pokemon = df[df["Type1"]=="Water"]
print(water_pokemon)
```

```python
# Filtering water type pokemons from Type 1 or Type 2 using OR (|) function

water_pokemon = df[(df["Type1"]=="Water") | (df["Type2"]=="Water")]
print(water_pokemon)
```

```python
# Filtering fire and flying type pokemons from Type 1 and Type 2 using AND (&) function

fire_flying_pokemon = df[(df["Type1"]=="Fire") & (df["Type2"]=="Flying")]
print(fire_flying_pokemon)
```

```python
# Filtering by specific columns ( Name and Type1 )

print(df[["Name", "Type1"]].to_string())
```

```python
# Display Name, Height and Weight of those Pokemons whose Height > 2
#              (Combine Row + Colum Filtering)

print(df.loc[df["Height"]>2, ["Name","Height","Weight"]])
```

```python
# Display all those pokemons whose Type1 has Fire or Water
#                      String Filering

print(df[df["Type1"].isin(["Fire","Water"])])
```

```python
# Display all those pokemons who does not have Fire as their Type1
#                     Negative Filtering using (~)


print(df[~df["Type1"].isin(["Fire","Water"])])
```

### Aggregate Functions

```python
# From whole dataframe

print(df.mean(numeric_only=True))   # Average
print(df.sum(numeric_only=True))    # Sum
print(df.max(numeric_only=True))    # Maximum
print(df.min(numeric_only=True))    # Minimum
print(df.count(numeric_only=True))  # Count
```

```python
# From selected column ( Height )

print(df["Height"].mean())
print(df["Height"].sum())
print(df["Height"].max())
print(df["Height"].min())
print(df["Height"].count())
```

```python
# Grouping using groupby()
# Grouping all similar types from Type1 ( ex : Bug , Dragon, etc) and finding mean , sum , max , min , count

group = df.groupby("Type1")
print(group["Height"].mean())
print(group["Height"].sum())
print(group["Height"].max())
print(group["Height"].min())
print(group["Height"].count())
```

### Data Cleaning

```python
# Drop irrelevant column(s)  -  Here dropping "Legendary" and "No" columns 

import pandas as pd
df = pd.read_csv("pokemon.csv")
df = df.drop(columns=["Legendary","No"])
print(df.to_string())
```

```python
# Handle Missing Data  -  Here dropping a row that is missing a value in Type2

df = df.dropna(subset=["Type2"])
print(df.to_string())
```

```python
# Replace missing values in rows with NONE 

import pandas as pd
df = pd.read_csv("pokemon.csv")
df = df.fillna({"Type2":"None"})
print(df.to_string())
```

```python
# Fix inconsistent values  -  Here replacing Grass in Type1 with GRASS

df["Type1"] = df["Type1"].replace({"Grass":"GRASS"})
print(df.to_string())
```

```python
# Standardize text   -   Converting all names of Pokemons to upper Case 

df["Name"]=df["Name"].str.upper()
print(df.to_string())
```

```python
# Fix data Types  -  Converting 0 & 1 (int) in Legendary to True & False (boolean)

df["Legendary"] = df["Legendary"].astype(bool)
print(df.to_string())
```

```python
# Removing Duplicate Values

df = df.drop_duplicates()
print(df.to_string())
```
